```python
sns.kdeplot(x='Age',data=df,fill=True,hue='Department')
plt.title("Age Density by Department")
```

```python
Attrition_count= df['Attrition'].value_counts()
plt.pie(Attrition_count, labels=Attrition_count.index, 
        autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2", len(Attrition_count)))
plt.title("Employee Attrition Distribution")
```

```python
AgeGroub_count= df['AgeGroup'].value_counts()
plt.pie(AgeGroub_count, labels=AgeGroub_count.index, autopct='%1.1f%%', startangle=90,
        colors=sns.color_palette("Set2", len(AgeGroub_count)))
plt.title("Employee Distribution by Age Group")
```

```python
BT_count= df['BusinessTravel'].value_counts()
plt.pie(BT_count, labels=BT_count.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2", len(BT_count)))
plt.title("Distribution of Business Travel Frequency Among Employees")
```

```python
sns.histplot(data = df, x ='BusinessTravel', hue = 'Gender' )
plt.title("Business Travel Frequency by Gender")
```

```python
sns.displot(data=df['JobRole'])
plt.xticks(rotation=90)
plt.title("Distribution of Employees by Job Role")
plt.show()
```

```python
SalarySlab_counts = df['SalarySlab'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(SalarySlab_counts, labels=SalarySlab_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Proportion of SalarySlab Usage', fontsize=10)
plt.axis('equal')  
plt.title("Proportion of Employees Across Different Salary Levels")
plt.show()
```

```python
satisfaction_counts = df.groupby('SalarySlab')['EnvironmentSatisfaction'].value_counts().unstack()
for slab in satisfaction_counts.index:
    plt.figure(figsize=(5, 5))
    plt.pie(
        satisfaction_counts.loc[slab], 
        labels=satisfaction_counts.columns, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=sns.color_palette('Set2')
    )
    plt.title(f'Environment Satisfaction Distribution for {slab} Salary Slab', fontsize=12)
    plt.axis('equal')  
    plt.show()
```

```python
df.groupby('JobRole').agg({'MonthlyIncome':'mean'}).reset_index().round({'MonthlyIncome': 1})
```

```python
pivot_table = df.pivot_table(
    values='NumCompaniesWorked',  
    index='EmpID', 
    aggfunc='sum'  
).rename(columns={'NumCompaniesWorked': 'TotalCompaniesWorked'})  # Rename column for clarity
top_10 = pivot_table.sort_values(by='TotalCompaniesWorked', ascending=False).head(10)
print(top_10)
```

```python
pivot_table = df.pivot_table(
    values='MonthlyIncome',    
    index='EducationField',  
    columns='Department',    
    aggfunc='mean',
    fill_value=0 
)


pivot_table = pivot_table.round(0)
pivot_table
```

```python
pivot_table = df.pivot_table(
    values='YearsWithCurrManager',    
    index='JobRole', 
    columns='Department',     
    aggfunc='mean',           
    fill_value=0                
)


pivot_table = pivot_table.round(1)
pivot_table
```

```python
pivot_table = df.pivot_table(
    values='JobInvolvement',    
    index='EducationField',      
    aggfunc=['mean', 'min', 'max'],            
    fill_value=0               
)
pivot_table = pivot_table.round(0)
pivot_table
```

```python
pivot_table = df.pivot_table(
    values='TotalWorkingYears',
    index='EmpID',              
    aggfunc='sum'                 
).rename(columns={'TotalWorkingYears': 'Total Years Worked'})  
top_10 = pivot_table.sort_values(by='Total Years Worked', ascending=False).head(10)
print(top_10)
```

```python
pivot_table = df.pivot_table(
    values='YearsAtCompany',   
    index='EmpID',             
    aggfunc='sum'             
).rename(columns={'YearsAtCompany': 'Total Years At Company'})  
top_10 = pivot_table.sort_values(by='Total Years At Company', ascending=False).head(10)
print(top_10)
```

```python
pivot_table = df.pivot_table(
    values='DistanceFromHome',  
    index='EmpID',   
    aggfunc='mean'        
).rename(columns={'DistanceFromHome': 'Count'}) 

top_10 = pivot_table.sort_values(by='Count', ascending=False).head(10)

print(top_10)
```
